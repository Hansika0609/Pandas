```python
import pandas as pd

# Load Excel file
df = pd.read_excel("youtube_comments.xlsx")

# Display first few rows
print(df.head())
```

```python
# Find out how many comments each user posted:
comment_counts = df["username"].value_counts()
print("Number of comments per user:\n", comment_counts)
```

```python
# Find the top 3 most liked comments:

top_liked = df.nlargest(3, "likes")[["username", "comment", "likes"]]
print("Top 3 most liked comments:\n", top_liked)

```

```python
from collections import Counter

# Convert comments to lowercase and split into words
all_words = " ".join(df["comment"].astype(str)).lower().split()
print(all_words)
# Count word frequency
word_counts = Counter(all_words)

# Show top 5 words
print("Most common words:\n", word_counts.most_common(5))
```

```python
df["comment_length"] = df["comment"].apply(len)
# print(df["comment_length"] )
longest_comment = df.loc[df["comment_length"].idxmax(), ["username", "comment"]]
shortest_comment = df.loc[df["comment_length"].idxmin(), ["username", "comment"]]

print("Longest comment:\n", longest_comment)
print("\nShortest comment:\n", shortest_comment)
```

```python
top_replied_users = df.groupby("username")["replies"].sum().sort_values(ascending=False)
print("Users with most replies:\n", top_replied_users)
```
